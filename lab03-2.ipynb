```python
import random
import math
import matplotlib.pyplot as plt
pi=math.pi
```

```python
n = 0 # number of points falling in the unit circle
d = 0 # number of points falling in the unit square
simulating=True

while simulating:
    x = random.random()
    y = random.random()
    if x**2 + y**2 <= 1.0:
        n += 1
    d += 1
    ratio = 4 * n * 1./d
    print(ratio)
    if abs(ratio-pi) / pi <= 0.01:
        print("Draws needed: ", d)
        break
```
